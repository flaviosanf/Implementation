```python
import numpy as np
import matplotlib.pyplot as plt
import ModuloFlavio as mod
import plt_conf as conf
import ModuloPlotTrian as mpt #Para importar la imagen de triangulo
import os
from scipy.interpolate import interp1d

conf.general()  # cargando configuración general

col = ['#1a1919', '#f0784d', '#2681ab', '#ab262f', '#bc92e0', '#486318', '#ed5b0c', 
      '#f0a092', '#484f07', '#694d0c','#0b5345','#154360']
```

```python
"Exportación de Datos SgrA"

C0=4.1e6
lambda_vals = np.logspace(-2,1,10)
lambda_vals=np.insert(lambda_vals,0,1e-5)

resultados_SgraASol = []
resultados_SgrACont = []

carpeta_destino = "Datos_SgrA"

### Numerical f(r) Solutions###
for lambda_val in lambda_vals:
    SgraASol, SgrACont = mod.NumSol(C0, lambda_val, Npt=50000, 
                                    iter='d', xmin=0, xmax=25, 
                                    metodo='RK45', Rtol=1e-9, Atol=1e-10, 
                                    info=True)
    resultados_SgraASol.append(SgraASol)
    resultados_SgrACont.append(SgrACont)

PairSolutions = [list(zip(resultado.t, resultado.y[0])) for resultado in resultados_SgraASol]

# Asegurarse de que la carpeta esté vacía
if os.path.exists(carpeta_destino):
    # Eliminar todos los archivos y subcarpetas en la carpeta
    for archivo in os.listdir(carpeta_destino):
        ruta_completa = os.path.join(carpeta_destino, archivo)
        if os.path.isfile(ruta_completa) or os.path.islink(ruta_completa):
            os.unlink(ruta_completa)  # Eliminar archivo o enlace simbólico

for i, pares in enumerate(PairSolutions):
    nombre_archivo = f"Sol_SgrA_{lambda_vals[i]:.4f}.dat"
    ruta_archivo = os.path.join(carpeta_destino, nombre_archivo)
    
    # Guardar los datos en un archivo .txt
    with open(ruta_archivo, 'w') as archivo:
        for t, y in pares:
            archivo.write(f"{t:.10f}\t{y:.10f}\n")

np.savetxt("lambda_Vals.dat", lambda_vals, fmt='%.8f')
```
